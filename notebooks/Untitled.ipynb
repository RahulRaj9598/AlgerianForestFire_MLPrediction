```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
##missing values
dataset[dataset.isnull().any(axis=1)]
```

### Dataset is converted into two sets based on Region from 122th index, we 
    can make a new column based on the region
### 1: "Bejaia Region"
### 2: "Sidi-Bel Abbes Region Dataset"
    Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column name
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.head()
```

```python
### change the required columns as integer datatype
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
## Encoding of the categories in classes
df_copy['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
## Percentage for piechart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```
